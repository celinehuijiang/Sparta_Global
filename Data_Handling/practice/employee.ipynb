```python
import pandas as pd
```

```python
df = pd.read_excel("Employee Sample Data.xlsx")
df
```

```python
# dropiing the column 
columns_to_drop = ['Exit Date']  
df_1 = df.drop(columns=columns_to_drop)
```

```python
df_1
```

```python
# checking for any of the duplicate 
num_duplicates = df_1.duplicated().sum()
num_duplicates
```

```python
#removing the duplicate 
no_duplicate = df_1.drop_duplicates()
no_duplicate

#checking if there is anymore duplicate 
no_duplicate.duplicated().sum()
```

```python
# checking the date is correcly ordered 
is_ordered = no_duplicate['Hire Date'].is_monotonic

if is_ordered:
    print("Dates are correctly ordered.")
else:
    print("Dates are not correctly ordered.")
```

```python
# sort the DataFrame by the "Date" column from oldesr to newest
ordered_df = no_duplicate.sort_values('Hire Date')

# reset the index
ordered_df = ordered_df.reset_index(drop=True)

# print the updated DataFrame
ordered_df
```

```python
# how to check for any white space 
whitespace_check = ordered_df['Job Title'].str.isspace()
whitespace_check.duplicated().sum()
```

```python
ordered_df = ordered_df
filename = 'clean.csv'
ordered_df.to_csv(filename, index=False)
```
