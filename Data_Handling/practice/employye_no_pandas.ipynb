```python
# reading the file
import openpyxl
```

```python
workbook = openpyxl.load_workbook('Employee Sample Data.xlsx')
sheet = workbook['Data']
print(sheet)
```

```python
# getting the list of row
rows = []  # Create an empty list to store the rows

for row in sheet.iter_rows(values_only=True):
    rows.append(list(row))

print(rows)
```

```python
# removing the duplicates
def remove_duplicates(lst):
    seen = set()
    for item in lst:
        if item in seen:
            return True
        seen.add(item)
    return False

print(remove_duplicates(rows))
```

```python
# check for any spaces within string
def has_whitespace(lst):
    for item in lst:
        if isinstance(item, str) and any(char.isspace() for char in item):
            return True
    return False

has_whitespace(rows)
```

```python
# removing the whitespace
# Assuming you have a 2D list called 'rows' containing the data

# Iterate over each row
for i in range(len(rows)):
    rows[i] = rows[i][:-1]  # Remove the last element from each row

print(rows)
```

```python
# Specify the file path for the new file
output_file = 'output.xlsx'

# Save the workbook to the specified file path
workbook.save(output_file)

print("Data saved to", output_file)
```
