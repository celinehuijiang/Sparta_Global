```python
import random

pokemon_data = {
    'name': 'Pikachu',
    'stats': [
        {'stat': {'name': 'hp'}, 'base_stat': 35},
        {'stat': {'name': 'attack'}, 'base_stat': 55},
        {'stat': {'name': 'defense'}, 'base_stat': 40}
    ],
    'abilities': [
        {'ability': {'name': 'static'}},
        {'ability': {'name': 'lightning-rod'}}
    ]
}

def pokemon_game():
    pokemon = random.choice(pokemon_data)
    name = pokemon['name']
    stats = [(stat['stat']['name'], stat['base_stat']) for stat in pokemon['stats']]
    abilities = [ability['ability']['name'] for ability in pokemon['abilities']]
    
    print("Welcome to the Pokémon Guessing Game!")
    print("I'm thinking of a Pokémon. Can you guess its name?")
    
    attempts = 0
    while True:
        guess = input("Enter your guess (or 'quit' to exit): ")
        attempts += 1
        
        if guess.lower() == 'quit':
            print("Thanks for playing! The Pokémon was", name)
            break
        
        if guess.lower() == name.lower():
            print("Congratulations! You guessed the Pokémon correctly!")
            print("Number of attempts:", attempts)
            break
        else:
            print("That's not the right Pokémon. Try again!")
            print("Here are some hints:")
            print("Stats:", stats)
            print("Abilities:", abilities)

# Call the pokemon_game function to start the game
pokemon_game()


```
